```python
store = df['priority_store'].value_counts()
store
```

```python
priority_store = df['priority_store'].value_counts().keys().tolist()
priority_store_count = df['priority_store'].value_counts().tolist()
store_zip = list(zip(priority_store,priority_store_count))
store_df= pd.DataFrame(data= store_zip, columns=['priority_store', 'priority_store_count'])
store_df

```

```python
priority_store_df = store_df[store_df['priority_store_count']>=7]
priority_store_df
```

```python
plt.bar(priority_store_df['priority_store'], priority_store_df['priority_store_count'])
plt.xticks(rotation = 45)
plt.show
```

```python
plt.pie(priority_store_df['priority_store_count'], labels = priority_store_df['priority_store'],autopct='%1.2f%%')
plt.show
```

```python
category_name = []
category_count  = []
for x in df.category:
    for i in range(len(x)):
        if x[i] in category_name: 
            category_count[category_name.index(x[i])]+=1 
        else: 
            category_name.append(x[i])
            category_count.append(1)

category = list(zip(category_name,category_count))
category_df = pd.DataFrame(data=category, columns=['category', 'num_of_products'])
category_df
```

```python
average_product = np.median(category_df['num_of_products']) +  np.max(category_df['num_of_products'])
average_product/2
```

```python
category_df=category_df[category_df['num_of_products']>average_product/2]
category_df
```

```python
plt.bar(category_df['category'], category_df['num_of_products'])
plt.xticks(rotation = 45)
plt.show
```

```python
plt.pie(category_df['num_of_products'], labels = category_df['category'],autopct='%1.2f%%')
plt.show
```
